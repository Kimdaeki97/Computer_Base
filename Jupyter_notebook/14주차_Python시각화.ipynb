# Python 시각화: Seaborn 2

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]  # 차트 사이즈 조정
```

## 1. Scatter Plot

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color=
               'g')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', 
                data=tips, 
                color='g',
                marker='x')
plt.show()
```

## 2. Seaborn 꾸미기

```python
# Grid(격자) 넣기
#sns.set_style('darkgrid')
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip', 
                data=tips, 
                color='g',
                marker='s')
plt.show()
```

```python
# paper, notebook, talk, poster
sns.set_context('notebook')
sns.scatterplot(x='total_bill', y='tip', 
                data=tips, 
                color='g',
                marker='s')
plt.show()
```

## 3. Cluster Maps

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

```python
pivot = flights.pivot_table(index='month', 
                    columns='year',
                    values='passengers')
pivot.head()
```

```python
sns.clustermap(pivot,
               cmap='coolwarm',
               linecolor='black',
               linewidth=2.0)
plt.show()
```

## 4. Pair Grids

```python
plt.rcParams['figure.figsize'] = [20, 12]
sns.pairplot(tips)
plt.show()
```

```python
pgrids = sns.PairGrid(tips)
pgrids.map(plt.scatter)
plt.show()
```

```python
pgrids.map_diag(sns.histplot)
pgrids.map_upper(sns.kdeplot)
pgrids.map_lower(plt.scatter)
plt.show()
```

## 6. Facet Grids

```python
tips.head()
```

```python
fgrid = sns.FacetGrid(data=tips, col='sex', row='time')
fgrid.map(sns.distplot, 'total_bill')
plt.show()
```

```python
fgrid = sns.FacetGrid(data=tips, col='sex', row='time')
fgrid.map(plt.scatter, 'total_bill', 'size')
plt.show()
```

## 6. Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', 
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           hue='sex',
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           hue='time',
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           hue='day',
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           hue='size',
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           hue='smoker',
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           col='smoker',
           data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', 
           col='size',
           data=tips)
plt.show()
```

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

```python
flights['passengers'].iplot()
```

## 2. Bar Plot

## 3. Scatter Plot

```python
flights.iplot(kind='scatter', 
              x= 'month', y='passengers',
              mode='markers',
              dash='solid')
```

```python
flights.iplot(kind='scatter', 
              x= 'month', y='passengers',
              mode='markers',
              color='green')
```
