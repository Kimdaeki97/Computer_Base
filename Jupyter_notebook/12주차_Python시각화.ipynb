## Matplotlib - Lineplot iris 실습

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 데이터 불러오기
iris = pd.read_csv('data/iris.csv')
#iris = pd.read_csv('./data/iris.csv')
iris.head()
```

```python
# 데이터의 기본 개요
iris.info()
```

### 1. Lineplot

```python
plt.figure(figsize = ([10, 8]))

plt.plot(iris['sepal_length'], iris['sepal_width'], c='g')

plt.title('Sepal Lineplot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

### 2. Scatter Plot

```python
plt.figure(figsize=([10, 8]))

plt.scatter(iris['sepal_length'], iris['sepal_width'], c='b', marker='x')
plt.scatter(iris['petal_length'], iris['petal_width'], c='r')

plt.title('Iris Scatter plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

### 3. Bar plot

```python
plt.bar(titanic['Sex'], titanic['Age'])

plt.title('Gender vs Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
plt.barh(titanic['Pclass'], titanic['Age'])

plt.title('Pclass vs Age')
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.show()
```

### 3. Histogram

```python
plt.hist(titanic['Age'])
plt.title('Age Histogram')
plt.show()
```

```python
plt.hist(titanic['Age'], bins=7)
plt.title('Age Histogram')
plt.show()
```

### 4.Pie Chart

```python
labels = ['IT', "Marketing", "Data Science", 'Finance']
values = [500, 156, 300, 510]
explode = [0.01, 0.01, 0.01, 0.1]

plt.pie(values, explode=explode, labels=labels, autopct='%1.2f%%')
plt.title('Classes Rate')

# 이미지 저장
plt.savefig('python01.jpg')

plt.show()
```

### 5. Sub Plot

```python
import math

plt.figsize = (12, 8)

x_vals = np.linspace(0, 20, 20)
y_vals = [math.sqrt(i) for i in x_vals]

plt.subplot(2, 2, 1)
plt.plot(x_vals, y_vals, 'bo-')

plt.subplot(2, 2, 2)
plt.plot(x_vals, y_vals, 'rx-')

plt.subplot(2, 2, 3)
plt.plot(x_vals, y_vals, 'g*-')

plt.subplot(2, 2, 4)
plt.plot(x_vals, y_vals, 'go-')

# 이미지 저장
plt.savefig('python02.jpg')

plt.show()
```
