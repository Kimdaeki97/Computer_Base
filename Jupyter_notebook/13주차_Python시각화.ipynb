# Python 시각화: Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head(3)
```

## 1.Histogram

```python
titanic['Age'].hist()
plt.show()
```

```python
titanic['Age'].hist(bins=20, color='orange')
plt.show()
```

## 2. Line Plots

```python
flights = pd.read_csv('data/flights.csv')
flights.head()
```

```python
flights.tail()
```

```python
flights.plot.line(y='passengers')
plt.show()
```

```python
flights.plot.line(x='year',
                  y='passengers',
                  color='orange')
plt.show()
```

## 3. Scatter Plots

```python
flights.head()
```

```python
flights.plot.scatter(x='year',
                     y='passengers',
                     color='red')
plt.show()
```

## 4. Bar Plots

```python
titanic['Sex'].value_counts().plot.bar(color='orange')
plt.show()
```

```python
titanic['Pclass'].value_counts().plot.bar(color='red')
plt.show()
```

```python
titanic['Pclass'].value_counts().plot.barh(color='red')
plt.title('BarH Chart')
plt.show()
```

## 5. Boxplot

```python
titanic.describe()
```

```python
titanic['Fare'].plot.box()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

## 1. Dist Plot

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

```python
tips.tail()
```

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.distplot(tips['tip'], kde=False)
plt.show()
```

```python
sns.histplot(tips['total_bill'])
plt.show()
```

# 2. Count plots

```python
titanic.head()
```

```python
sns.countplot(x='Pclass', data=titanic)
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Sex', data=titanic)
plt.show()
```

## 3. Joint Plot

```python
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.show()
```

```python
sns.jointplot(x='size', y='total_bill', data=tips)
plt.show()
```

## 4. Pairplot

```python
tips.describe()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, hue='day')
plt.show()
```

## 5. Heatmap

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.corr()
```

```python
corr = titanic.corr()

sns.set_context('paper')
sns.heatmap(corr, annot=True)
sns.heatmap(corr)
```

```python
ax = sns.heatmap(corr, annot=True, cmap='coolwarm')
```

## 6. Bar Plot

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
plt.show()
```

```python
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic)
plt.show()
```

```python
sns.catplot(x='Pclass', 
            y='Age', 
            hue='Sex', 
            col='Survived', 
            kind='bar',
            data=titanic)
plt.show()
```

## 7. Box Plot

```python
titanic.head()
```

```python
sns.boxplot(data=titanic)
```

```python
sns.boxplot(y=titanic['Age'])
```

```python
sns.boxplot(x='Survived', y='Age', data=titanic)
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', hue='Sex', data=titanic)
plt.show()
```

## 8. Violin Plot

```python
sns.violinplot(x='Survived', y='Age', data=titanic)
plt.show()
```

```python
sns.violinplot(x='Survived', y='Age', hue='Sex', data=titanic)
plt.show()
```

```python
sns.violinplot(x='Survived', y='Age', hue='Sex', data=titanic, split=True)
plt.show()
```

## 9. Strip Plot

```python
sns.stripplot(x='Survived', y='Age', data=titanic)
```

```python
sns.stripplot(x='Survived', y='Age', hue='Sex', data=titanic, split=True)
plt.show()
```

## 10. Swarm Plot

```python
sns.swarmplot(x='Survived', y='Age', data=titanic)
```

```python
sns.swarmplot(x='Survived', y='Age', hue='Sex', data=titanic, split=True)
plt.show()
```
